```python
dataset_m['Date']=pd.to_datetime(dataset_m['Date'])
train_df['Date']=pd.to_datetime(train_df['Date'])
```

```python
df=train_df.merge(dataset_m,how='inner',on=['Store','Date','IsHoliday'])
```

```python
df
```

```python
df['week']=df.Date.dt.isocalendar().week
```

```python
df['year']=df.Date.dt.isocalendar().year
```

```python
##isovisualization or eda(Exploratory Data Analysis)
##make a function which will make a scatter plot and take arguments of df and column which we want to plot again
def plot(df,column):
    plt.figure(figsize=(12,8))
    plt.scatter(df[column],df['Weekly_Sales'])
    plt.xlabel(column)
    plt.ylabel('Weekly_Sales')
```

```python
plot(df,'Store')
```

```python
plot(df,'Dept')
```

```python
df.year.value_counts()
```

```python
##in year 2011 average weekly sales using a line plot
df['Weekly_Sales'].info()
```

```python
weekly_sales_2011=df[df['year']==2011].groupby('week')['Weekly_Sales'].mean()
```

```python
plt.plot(weekly_sales_2011.index,weekly_sales_2011.values,color='r')
plt.xlabel('week')
```

```python
weekly_sales_2010=df[df['year']==2010].groupby('week')['Weekly_Sales'].mean()
```

```python
plt.plot(weekly_sales_2010.index,weekly_sales_2010.values)
plt.xlabel('week')
```

```python
weekly_sales_2012=df[df['year']==2012].groupby('week')['Weekly_Sales'].mean()
```

```python
plt.plot(weekly_sales_2012.index,weekly_sales_2012.values,color='g')
plt.xlabel('week')
```

```python
##combile plot of year 2010,2011,2012
plt.figure(figsize=(20,10))
plt.plot(weekly_sales_2010.index,weekly_sales_2010.values)
plt.plot(weekly_sales_2011.index,weekly_sales_2011.values,color='r')
plt.plot(weekly_sales_2012.index,weekly_sales_2012.values,color='g')
plt.title('Average Weekly Sales per Year',fontsize=30)
plt.grid()
plt.xlabel('week',fontsize=20)
plt.ylabel('Sales',fontsize=20)
plt.legend(['2010','2011','2012'])
plt.xticks(np.arange(1,60,step=1))
plt.show()
```
